```python
#Importing the necessary libraries for data manipulation and analysis

import pandas as pd
```

```python
# Reading the CSV file into a dataframe
file_path = 'C:/Users/Lenovo/Downloads/Iris.csv'
iris_df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
iris_df.head()
```

```python
# Selecting petal length and sepal length
selected_df = iris_df[['PetalLengthCm', 'SepalLengthCm']]

# Display the first few rows of the selected dataframe
selected_df.head()
```

```python
# Filtering data for flowers with petal length more than 5.5 units
filtered_df = iris_df[iris_df['PetalLengthCm'] > 5.5]

# Display the first few rows of the filtered dataframe
filtered_df.head()
```

```python
# Creating the new column ‘flower area’
filtered_df['flower_area'] = filtered_df['PetalLengthCm'] * filtered_df['PetalWidthCm']

# Display the first few rows to check the new column
filtered_df.head()
```
